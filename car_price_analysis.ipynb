```python
import pandas as pd
import numpy as np
```

```python
data_source="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
```

```python
headers=["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df1=pd.read_csv(data_source,names=headers)
df1.head(20)
```

```python
df1.info()
```

```python
df1.replace("?",np.nan,inplace=True)
df1.info()
```

```python
# counting how many missing values are there in each columns
df1.isnull()
```

```python
missing_data=df1.isnull()
missing_data.head(15)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print(" ")
```

```python
# Dealing with missing data
```

```python
df1['normalized-losses']=df1['normalized-losses'].astype('float')
df1.replace(np.nan,{'normalized-losses':df1['normalized-losses'].mean()},inplace=True)
df1['normalized-losses'].info()
```

```python
df1['bore']=df1['bore'].astype('float')
df1.replace(np.nan,{'bore':df1['bore'].mean()},inplace=True)
df1['bore'].info()
```

```python
df1[['stroke','horsepower','peak-rpm']]=df1[['stroke','horsepower','peak-rpm']].astype('float')
```

```python
df1.replace(np.nan,{'stroke':df1['stroke'].mean(),'horsepower':df1['horsepower'].mean(),'peak-rpm':df1['peak-rpm'].mean()},inplace=True)
df1.info()
```

```python
df1['num-of-doors'].info()
```

```python
df1['num-of-doors'].value_counts()
```

```python
max_value=df1['num-of-doors'].value_counts().idxmax()
max_value
```

```python
df1.replace(np.nan,{'num-of-doors':'four'},inplace=True)
df1['num-of-doors'].info()
```

```python
df1.dropna(subset=['price'],axis=0,inplace=True)
```

```python
df1.reset_index(drop=True,inplace=True)
```

```python
# Correcting the data format
```

```python
df1['price']=df1['price'].astype('float')
df1['normalized-losses']=df1['normalized-losses'].astype('int')
```

```python
df1.isnull()
```

```python
# Data Standarization
```

```python
# Convert city-mpg to city-l/100 km
```

```python
df1['city-L/100 Km']=235/df1['city-mpg']
df1['city-L/100 Km'].head()
```

```python
# convert highway-mpg to highway-L/100 km
```

```python
df1['highway-L/100 Km']=235/df1['highway-mpg']
df1['highway-L/100 Km'].head()
```

```python
# Data Normalization
```

```python
df1[['length','width','height']].head()
```

```python
df1['length']=df1['length']/df1['length'].max()
df1['width']=df1['width']/df1['width'].max()
df1['height']=df1['height']/df1['height'].max()
```

```python
df1[['length','width','height']].head()
```

```python
# Concept of Binning
```

```python
df1['horsepower']=df1['horsepower'].astype(int,copy=True)
```

```python
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df1['horsepower'])
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('counts')
plt.pyplot.title('horsepower histrogram')
```

```python
df1['horsepower-binned'].value_counts()
```

```python
# plotting the distribution of each bins
pyplot.bar(group_names,df1['horsepower-binned'].value_counts())
pyplot.xlabel('horsepower')
pyplot.ylabel('counts')
pyplot.title('horsepower bins visulaization')
```

```python
plt.pyplot.hist(df1["horsepower"], bins = 3)
```

```python
df1['fuel-type'].value_counts()
```

```python
dummy_variable_1=pd.get_dummies(df1['fuel-type'])
dummy_variable_1
```

```python
df1=pd.concat([df1,dummy_variable_1],axis=1)
df1.head()
```

```python
df1.drop('fuel-type',axis=1,inplace=True)
df1.head()
```

```python
df1['aspiration'].value_counts()
```

```python
dummy_variable_2=pd.get_dummies(df1['aspiration'])
dummy_variable_2.head()
```

```python
df1=pd.concat([df1,dummy_variable_2],axis=1)
df1.head(10)
```

```python
df1.drop('aspiration',axis=1,inplace=True)
```

```python
# ANOVA
```

```python
group_test3=df_group_one[['drive-wheels','price']].groupby(['drive-wheels'])
group_test3.head()
```

```python
group_test3.get_group(('4wd',))['price']
```

```python
F_value,P_value=stats.f_oneway(group_test3.get_group(('4wd',))['price'],group_test3.get_group(('fwd',))['price'],group_test3.get_group(('rwd',))['price'])
print(F_value,P_value)
```

```python
# 4wd vs rwd
f_value,p_value=stats.f_oneway(group_test3.get_group(('4wd',))['price'],group_test3.get_group(('rwd',))['price'])
print(f" The F-statistics value is {f_value} and the p-value is {p_value}")
```

```python
# 4wd vs fwd
f_value,p_value=stats.f_oneway(group_test3.get_group(('4wd',))['price'],group_test3.get_group(('fwd',))['price'])
print(f" The F-statistics value is {f_value} and the p-value is {p_value}")
```

```python
# rwd vs fwd
f_value,p_value=stats.f_oneway(group_test3.get_group(('rwd',))['price'],group_test3.get_group(('fwd',))['price'])
print(f" The F-statistics value is {f_value} and the p-value is {p_value}")
```
